```python
df = pd.read_csv(data_path, usecols=['Close'])
```

```python
df = df[:700]
```

#### MACD

```python
ewma = pd.Series.ewm

df['close_26EMA'] = ewma(df["Close"][25:], span=26).mean()
df['close_12EMA'] = ewma(df["Close"][11:], span=12).mean()

df['MACD'] = df['close_12EMA'] - df['close_26EMA']
df['MACD_9EMA'] = ewma(df["MACD"][8:], span=9).mean()

df = df.fillna(0)
```
